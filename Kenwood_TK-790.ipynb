# Kenwood TK-790

The Kenwood TK-x90 radios series is a discontinued land-mobile radio (FCC part 90) that was used extensively in public safety (police/fire) in the US.  These are rugged, mil-spec quality radios.  These radios are being phased out in lieu of digital radios and is currently available in the second-hand market in the US for $50-80, depending on condition.  The VHF models can be converted to work on the 2m band, making them an inexpensive option for a dedicated APRS radio. Compared to the Motorola SM50 reviewed earlier, these are much more rugged and capable radios.

These radios are also 9600-baud capable, making them useful for setting up 9600 baud packet stations.

There are a number of reasons that a radio of this quality sells as cheaply as it does.  It presents both challenges and an opportunity.  The goal of this article is to go through the challenges with these radios, and how to overcome them.

We will also goes over the features of the radio, how to program it for the 2m ham band, how to set it up as a dedicated APRS/packet radio, and what is needed to make it work at 9600 baud.

## Assumptions

This article assumes you can read and understand basic schematic diagrams, have a soldering iron, can build your own cables, and can create basic circuits on a breadboard.  We will also be using a 3D printer to create a custom connector.  Advanced users may build a programmer using a custom PCB with surface-mount components.

We will be using Linux and Wine to run the programming software to program the radio.

The article assumes you will be connecting the radio to a Mobilinkd TNC3, and have another APRS-capable transceiver for testing. You should also have a dummy load capable of 25W or more, and an RTL SDR.

## Introduction

After exploring the SM50 and being unable to get it to work at 9600 baud, I decided to try a radio that was much more likely to work at 9600 baud.  Reading the specs in the service manual convinced me that, while not guaranteed, that it would very likely work.  And it would certainly work at 1200 baud, still being good value for a dedicated APRS radio.

However, one of the bigger challenges is that the connector used for the microphone and programmer is decidedly proprietary to Kenwood, and apparently only ever used on this series of radios.  Because of this, used microphones in poor condition fetch a premium.  Even an after-market programmer available for this radio fetches a premium at $30-40, more than the radio itself can sometimes be found for.

I felt that these issues could be overcome with a 3D printer and rudimentary modeling skills.

### Models

There are three major models each with multiple variants.  There are also normal (45W) and "H" (100W) variants.  We will go into these in detail below.

The major models to be aware of are:

 * TK-690 -- VHF low-band (29.7-50MHz), three variants, generally considered unusable for ham bands
 * TK-790 -- VHF high-band (136-174MHz), two variants, usable on 2m
 * TK-890 -- UHF (403-512MHz), three variants, only one known usable on 70cm
 
We will not be covering the "H" variants here.  The majority of these are sold without a control head, and adding a control head and cable can cost more than the radio.  There are a lot of traps one can fall in with these radios, and covering them all would take a complete article in itself.  Let's just say that if you buy an "H" model, do not be surprised to find that you have an inert piece of metal on your hands.

There are also two control heads available for these units.  They are often sold as separate items (primarily for the "H" variants).

    The first trap to avoid when buying one of these radios is making sure you
    are buying a full radio with control head and not just the control head.

### Control Heads

There are two control heads available for these radios, the KCH-10 basic control head, and the KCH-11 advanced control head.  The major difference is that the basic control head has a built-in speaker and fewer buttons, while the advanced control head has a lot more buttons but does not have a speaker.

#### Basic Control Head KCH-10

The basic control is probably the most useful for amateur radio use as it has a built-in speaker.  My experience is that these are quite a bit less common than the KCH-11.  However, one of the more noticeable limitations of this control head is that it can display fewer characters on its display.

#### Advanced Control Head KCH-11

The advanced control head is the more common control head found on these radios.  These have a larger display screen and 6 more buttons than the KCH-10.  To get the added buttons and screen size, it gives up the speaker.

If you want speaker output from the radio, you will need to connect an external speaker to the 9-pin Molex connector on the rear of the radio.
